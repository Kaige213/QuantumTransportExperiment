# 一、数字滤波器


Note：

 - 可以看出，滤波器有很明显的相移；
 
 - 需要高阶滤波器时，sos filter的数值稳定性更好，更加通用；

````python
#生成带噪声的信号
def signal_func(t_vec, freq, amplitude, phase, noise_amplitude):
    signal_vec = amplitude*np.sin(2*np.pi*freq*t_vec + phase/180*np.pi)
    noise_vec  = np.random.normal(0, noise_amplitude, len(t_vec))
    return signal_vec + noise_vec
````

## 1.1 数字低通滤波器

``` python
#数字低通滤波器的使用
LP_order   = 8
LP_cutoff  = 150
#butter filter
lp_b, lp_a = signal.butter(LP_order, LP_cutoff, 'low', fs=fs)       
signal_out = signal.lfilter(lp_b, lp_a, signal_in)
#sos filter
sos         = signal.butter(LP_order, LP_cutoff, btype='low', fs=fs, output='sos')
signal_out  = signal.sosfilt(sos, signal_in)

```

````python
freq      = 100
fs        = 10001
T         = 1 / freq
t_end     = 10*T
t_vec     = np.linspace(0,t_end, int(t_end*fs+1))
amplitude = 1
phase     = 45
noise_amplitude = 1
signal_vec      = signal_func(t_vec, freq, amplitude, phase, noise_amplitude)
signal_noNoise_vec = signal_func(t_vec, freq, amplitude, phase, 0)

#低通滤波
LP_order   = 8
LP_cutoff  = 150
#butter filter
lp_b, lp_a = signal.butter(LP_order, LP_cutoff, 'low', fs=fs)
signal_out = signal.lfilter(lp_b, lp_a, signal_vec)

#==== plot ====
plt.plot(t_vec/T, signal_vec, c='C1')
plt.plot(t_vec/T, signal_noNoise_vec, c='C2', label='Signal')
plt.plot(t_vec/T, signal_out, '--', label='Filter', c='C0')

plt.xlim([0, t_vec[-1]/T])
plt.xlabel("t / T", fontsize=13)
plt.ylabel('Voltage', fontsize=13)
plt.legend(fontsize=13)

plt.tight_layout()
#plt.savefig("pic_name.png", format='png', dpi=200)
````

# 二、数字锁相放大器

带有噪声的交流信号：
$$
V(t) = A\sin (\Omega t + \phi) + \delta V(t)
$$
其中，$\delta V(t)$表示噪声

与参照信号相乘：
$$
\begin{cases}
V(t) \times 2\sin(\Omega t) = A\cos \phi - A\cos(2\Omega t + \phi) + 2\delta V(t) \sin(\Omega t)
\\
V(t) \times 2\cos(\Omega t) = A\sin \phi + A\sin(2\Omega t + \phi) + 2\delta V(t) \cos(\Omega t)
\end{cases}
$$

上面的操作可以看成是对数据做了一次映射，映射结果的组成：直流 + 交流 。其中，直流是我们的目标值，而交流部分包含噪声，倍频（通常信号源有直流分量，所以与参照信号相乘后，形成新的基频成分）。最后，通过求平均，除去交流部分，留下直流分量。从这个角度来看，锁相放大器能测量弱信号，依然是根据“多次实验求平均”这一基本原理（相当朴素啊）。

通常，锁相放大器还会将上面的结果通过低通滤波器，然后才输出结果。如果是数字锁相放大器，其实都可以不用进行滤波，直接将上面与参照信号相乘的结果进行求平均，输出结果。因为低通滤波器不会影响数据中的直流分量，该直流分量在上面的乘法操作后就被确定了，无法改变。当然，如果是模拟锁相放大器，是必须要低通滤波器的，其作用就是对数据求平均，并方便最终读数。至于为什么数字锁相放大器要使用低通滤波器，可能是有其他方面的考虑，比如输出数字的变化可以直观反映待测频率附近的噪声大小。又或者这样的锁相放大器可以等效为带有放大效果的带通滤波器？

## 2.1 生成带噪声的电压信号

````python
#生成带噪声的信号
def signal_func(t_vec, freq, amplitude, phase, noise_amplitude):
    signal_vec = amplitude*np.sin(2*np.pi*freq*t_vec + phase/180*np.pi)
    noise_vec  = np.random.normal(0, noise_amplitude, len(t_vec))
    return signal_vec + noise_vec
````

````python
#生成带噪声的信号
fs        = 1001  #采样率
tf        = 1     #采样时长
t_vec     = np.linspace(0,tf,fs*tf+1)
freq      = 100
amplitude = 1
phase     = 45
noise_amplitude = 1

signal_vec = signal_func(t_vec, freq, amplitude, phase, noise_amplitude)

#==== plot ====
plt.plot(t_vec, signal_vec)
plt.xlim([t_vec[0], t_vec[-1]])
plt.xlabel('Time', fontsize=13)
plt.ylabel('Voltage', fontsize=13)
````

## 2.2 相乘后直接平均

````python
#与参照函数相乘，然后平均
X = np.average(2*signal_vec*np.sin(2*np.pi*t_vec*freq))
Y = np.average(2*signal_vec*np.cos(2*np.pi*t_vec*freq))


print('R: ', np.abs(X+1j*Y))
print('Phase: ', np.angle(X+1j*Y)*180/np.pi)
````

## 2.3 相乘+滤波

````python
#low pass filter
order  = 4
cutoff = 20
b,a    = signal.butter(order, cutoff, 'low', fs=fs) 

sX  = 2*signal.lfilter(b,a,signal_vec*np.sin(2*np.pi*t_vec*freq))  #相乘并滤波
sY  = 2*signal.lfilter(b,a,signal_vec*np.cos(2*np.pi*t_vec*freq))  #相乘并滤波

#average
X  = np.average(sX)
Y  = np.average(sY)

print('R: ', np.abs(X+1j*Y))
print('Phase: ', np.angle(X+1j*Y)*180/np.pi)

#==== plot ====
plt.figure(figsize=(9,3))

plt.subplot(1,2,1)
plt.plot(t_vec, sX)
plt.xlim([t_vec[0], t_vec[-1]])
plt.xlabel('Time', fontsize=13)
plt.ylabel('sX', fontsize=13)


plt.subplot(1,2,2)
plt.plot(t_vec, sY)
plt.xlim([t_vec[0], t_vec[-1]])
plt.xlabel('Time', fontsize=13)
plt.ylabel('sY', fontsize=13)
````

# 三、模拟实验

````python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from tqdm import tqdm
````

## 3.1 用锁相放大器测试数字低通滤波器

````python
def Lockin_XY_func(signal_vec, fs, freq, cutoff=100, order=4, average=False):
    t_vec = np.linspace(0, len(signal_vec)/fs,len(signal_vec))      #时间坐标
    b, a  = signal.butter(order, cutoff, 'low', fs=fs)              #低通滤波器
    sX    = 2*signal.lfilter(b,a,signal_vec*np.sin(2*np.pi*t_vec*freq))  #相乘并滤波
    sY    = 2*signal.lfilter(b,a,signal_vec*np.cos(2*np.pi*t_vec*freq))  #相乘并滤波
    if (average == True):
        sX = np.sum(sX) / len(sX)
        sY = np.sum(sY) / len(sY)
    #如果不平均，则直接输出滤波后的信号
    return sX,sY

def signal_func(t_vec, freq):
    return np.sin(2*np.pi*freq*t_vec) #默认输入信号：amplitude=1, phase=0, noise=0

def lowPass_filter(signal_in, order, cutoff, fs):
    sos         = signal.butter(order, cutoff, btype='low', fs=fs, output='sos')
    signal_out  = signal.sosfilt(sos, signal_in)
    return signal_out
````

````python
#设置低通滤波器
order  = 4
cutoff = 100   

#扫描的频率范围
freq_vec = np.linspace(10,1000, 1001)
R_vec    = np.zeros(freq_vec.size)
phase_vec= np.zeros(freq_vec.size)

for n_freq in tqdm(range(freq_vec.size)):
    #生成信号
    freq  = freq_vec[n_freq]
    T     = 1 / freq
    #确定采样率，
    if (freq < cutoff): fs = 10*cutoff #如果信号的频率太小，则根据待测滤波器的截止频率来采样
    else:               fs = 10*freq       #一周期内，采样10个点
    t_end = 100*T
    t_vec = np.linspace(0, t_end, int(fs*t_end)+1)
    signal_vec = signal_func(t_vec, freq)
    #周期信号通过滤波器
    #sos filter
    signal_vec = lowPass_filter(signal_vec, order, cutoff, fs)
    
    #用锁相放大器测量，出射信号
    [sX, sY] = Lockin_XY_func(signal_vec, fs, freq, cutoff=freq/2, order=4, average=True)
    #记录数据
    R_vec[n_freq]     = np.abs(sX + 1j*sY)
    phase_vec[n_freq] = np.angle(sX + 1j*sY)*180/np.pi
````

````python
plt.semilogx(freq_vec, 20*np.log10(R_vec))
plt.title('Butterworth filter frequency response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(cutoff, color='green') # cutoff frequency
````

## 3.2 用锁相放大器测试数字带通滤波器

````python
def Lockin_XY_func(signal_vec, fs, freq, cutoff=100, order=4, average=False):
    t_vec = np.linspace(0, len(signal_vec)/fs,len(signal_vec))      #时间坐标
    b, a  = signal.butter(order, cutoff, 'low', fs=fs)              #低通滤波器
    sX    = 2*signal.lfilter(b,a,signal_vec*np.sin(2*np.pi*t_vec*freq))  #相乘并滤波
    sY    = 2*signal.lfilter(b,a,signal_vec*np.cos(2*np.pi*t_vec*freq))  #相乘并滤波
    if (average == True):
        sX = np.sum(sX) / len(sX)
        sY = np.sum(sY) / len(sY)
    #如果不平均，则直接输出全部的滤波后的信号
    return sX,sY

def signal_func(t_vec, freq):
    return np.sin(2*np.pi*freq*t_vec) #默认信号：amplitude=1, phase=0, noise=0

def bandPass_filter(signal_in, order, lowcut, highcut, fs):
    sos        = signal.butter(order, [lowcut, highcut], btype='band', fs=fs, output='sos')
    signal_out = signal.sosfilt(sos, signal_in)
    return signal_out
````

````python
#设置带通滤波器
order   = 9
lowcut  = 100
highcut = 200

#扫描的频率范围
freq_vec = np.linspace(10,1000, 1001)
R_vec    = np.zeros(freq_vec.size)
phase_vec= np.zeros(freq_vec.size)

for n_freq in tqdm(range(freq_vec.size)):
    
    #生成信号
    freq  = freq_vec[n_freq]
    T     = 1 / freq
    #确定采样率，
    if (freq < highcut): fs = 10*highcut #如果信号的频率太小，则根据滤波器确定采样频率
    else:                fs = 10*freq       #一周期内，采样10个点
    #采样点
    t_end = 100*T
    t_vec = np.linspace(0, t_end, int(fs*t_end)+1)
    signal_in = signal_func(t_vec, freq)
    
    #周期信号通过滤波器
    signal_out = bandPass_filter(signal_in, order, lowcut, highcut, fs)
    
    #用锁相放大器测量，出射信号
    [sX, sY] = Lockin_XY_func(signal_out, fs, freq, cutoff=freq/2, order=4, average=True)
    #记录数据
    R_vec[n_freq]     = np.abs(sX + 1j*sY)
    phase_vec[n_freq] = np.angle(sX + 1j*sY)*180/np.pi
````

````python
plt.semilogx(freq_vec, 20*np.log10(R_vec))
plt.title('Butterworth filter frequency response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(lowcut, color='green') 
plt.axvline(highcut, color='green')
plt.show()
````

### 3.3 数据处理时，直接平均与滤波后再平均？

如果信号的组成是直流+噪声：$\bar{V} + \delta V$，用采集卡测量到了一组值$[V_n, ...]$，直接平均的结果与将该组数据通过低通滤波器后再平均，后一种处理方式对$\bar{V}$的测量不会提高准确性。因为数字低通滤波器不会改变$[V_n, ...]$中的直流分量，只是将该组数据的分布向均值靠拢，看起来噪声变小，但均值不会变化。所以这种情况下，使用数字滤波器的测量方法并不能减小测量误差。

结论：
 - 上述情况下，只有增加测量次数，才能提高精度，使测量结果向真实值逼近。
 
 - 数字滤波器只能影响交流部分的信号，不能改变直流的精度


````python
def signal_func(sampleNum, noise_amplitude):
    return np.random.normal(0, noise_amplitude, sampleNum) #高斯噪声

Noise_amplitude    = 10
sampleNum_vec      = 2**np.arange(10,18)
varianceDirect_vec = np.zeros(sampleNum_vec.size)
varianceFilter_vec = np.zeros(sampleNum_vec.size)
N_exp              = 1000

for n_sampleNum in tqdm(range(sampleNum_vec.size)):
    sampleNum   = sampleNum_vec[n_sampleNum]
    #重复实验，每次实验中，采集sampleNum个数据点，然后平均
    expDirect_vec = np.zeros(N_exp)
    expFilter_vec = np.zeros(N_exp)
    for n_exp in range(N_exp):
        sample_vec = signal_func(sampleNum, Noise_amplitude)
        #先经过低通滤波器
        fs         = 1000
        LP_order   = 8
        LP_cutoff  = 100
        lp_b, lp_a = signal.butter(LP_order, LP_cutoff, 'low', fs=fs)       
        sample_out = signal.lfilter(lp_b, lp_a, sample_vec)
        
        #
        expDirect_vec[n_exp]  = np.sum(sample_vec) / sample_vec.size
        expFilter_vec[n_exp]  = np.sum(sample_out) / sample_out.size
    #计算1000次实验结果的分布的方差
    #没有滤波
    average = np.sum(expDirect_vec) / expDirect_vec.size
    varianceDirect_vec[n_sampleNum] = np.sum( (expDirect_vec-average)**2) / expDirect_vec.size
    #滤波结果
    average = np.sum(expFilter_vec) / expFilter_vec.size
    varianceFilter_vec[n_sampleNum] = np.sum( (expFilter_vec-average)**2) / expFilter_vec.size

plt.plot(np.log2(sampleNum_vec), np.log2(varianceDirect_vec), label='Without')
plt.plot(np.log2(sampleNum_vec), np.log2(varianceFilter_vec), '+', label='Filter')

plt.legend(fontsize=13)
plt.xlabel(r"$\rm log_2(N)$", fontsize=13)
plt.ylabel(r"$\rm log_2(Variance)$", fontsize=13)
````
