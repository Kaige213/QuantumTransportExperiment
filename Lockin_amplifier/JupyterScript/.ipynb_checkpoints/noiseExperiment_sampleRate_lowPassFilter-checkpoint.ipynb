## 1. 带噪声的电压信号

为了模拟噪声测量实验，我们使用函数voltageAfterFilter_timeDomain_func()用来生成随机电压信号，其数学表达式：
$$
  V(t) = \sum_{n=1}^{N_A} A_n \sin(n\Omega_f t) + \sum_{n=1}^{N_A} B_n \cos(n\Omega_f t)
$$

其中：
$$
    \begin{cases}
        \Omega_f = \frac{2\pi}{t_f}
        \\
        N_A = \mathrm{Int}[t_f f_0]
    \end{cases}
$$

$t_f$为采样时长，$f_0$为低通滤波器的截止频率。上面求和指标的上限$N_A$是由低通滤波器的截止确定，即生成的电压信号中没有高于该频率的成分。
为了保证该信号的平均功率不随参数$t_f$变化，随机系数$A_n$和$B_n$为 $\mu=0, \ \sigma^2=1/N_A$ 的高斯分布。

## 2. PDS

采集卡测量到离散的电压信号$V$（向量，存放电压数据）后，使用FFT来计算电压噪声：
$$
 S_V = \frac{t_f}{N_t^2} \big| \mathrm{FFT}(V) \big|^2
$$

为了后面推导方便，将上一节的电压函数写为：
$$
V(t) = \sum_{n=-N_A}^{N_A} A_n e^{in\Omega t}
$$

将电压函数代入噪声计算式中，得到：
$$
    S_V(f_k = k/t_f) = t_f \big| \sum_{n=-N_A}^{N_A} A_n \delta(n-k,lN_t)   \big|^2, \ \ l=0,\pm1, \pm2, ...
$$

而我们的目标是：$S_V(f_k) = t_f |A_n|^2$，即对应上面量$l=0$的部分。通过分析$N_t<2N_A$时，$l\neq0$的部分会产生影响，导致测量结果错误。所以在选择采样率时，应该满足：$ f_s > 2 f_0$

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
def voltageAfterFilter_timeDomain_func(t_vec):
    tf          = t_vec[-1] - t_vec[0]
    omega_f     = 2*np.pi/tf
    f0          = 100   
    NA          = round(float(np.floor(tf*f0)))
    voltage_vec = np.zeros(t_vec.size)
    noise_amplitude = 1/np.sqrt(NA)
    
    for n in range(1,NA+1):
        voltage_vec = voltage_vec + np.random.normal(0, noise_amplitude)*np.sin(n*omega_f*t_vec)
    for n in range(1,NA+1):
        voltage_vec = voltage_vec + np.random.normal(0, noise_amplitude)*np.cos(n*omega_f*t_vec)
    return voltage_vec

def noiseSpectrum_func(t_vec, voltage_vec):
    tf = t_vec[-1]-t_vec[0]
    voltageFreq_vec = np.fft.fft(voltage_vec)[1:(t_vec.size//2+1)]
    noiseSpectrum   = 2*np.abs(voltageFreq_vec)**2*tf/t_vec.size**2
    return noiseSpectrum
```

### 2.1 采样率 fs=80 sa/s 时

```python
Nt     = 80
tf     = 1
t_vec  = np.linspace(0, tf, Nt, endpoint=False)
freq_vec = 1/tf*np.arange(1, Nt//2+1)
Sw_vec   = np.zeros(freq_vec.size)

N_exp = 1000
for n_exp in tqdm(range(N_exp)):
    voltage_vec = voltageAfterFilter_timeDomain_func(t_vec)
    Sw_vec      = Sw_vec + noiseSpectrum_func(t_vec, voltage_vec) / N_exp
```

```python
plt.plot(freq_vec, Sw_vec)
plt.xlabel('f (Hz)', fontsize=13)
plt.ylabel('Noise', fontsize=13)
plt.xlim([0, freq_vec[-1]])
```
