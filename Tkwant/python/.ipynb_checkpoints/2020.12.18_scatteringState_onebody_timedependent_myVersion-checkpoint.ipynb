**说明：**

 - 虽然这里的归一化系数没有作用，但为了表达式统一，依然将它加到了程序中
 
 - 增加solve_ivp的精度，计算结果中，波函数的反常波纹消失
 
 - 使用sparse matrix，内存更小，计算速度更快

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.sparse as sp
from scipy.sparse.linalg import spsolve
from scipy.integrate import solve_ivp
from scipy.special import erf
```

无量纲化：$\frac{\hbar^2}{2ma^2}=1$, a= 1

色散关系： $E(k) = 2\frac{\hbar^2}{2ma^2} (1-\cos(ka))$

```python
#方势垒
def potential_func(num_total):
    V0 = 0
    L  = 20
    pos_start     = int(num_total/2-L/2) 
    potential_vec = np.zeros(num_total)
    potential_vec[ pos_start:(pos_start+L) ] = V0
    return potential_vec
```

```python
# 虚势的参数是根据tkwant得到的
num_lead_cells   = 88    #虚势区间长度
num_buffer_cells = 101    #用于缓冲长波的区间长度
num_center_cells = 400    #散射区长度
num_total_cells  = num_lead_cells*2 + num_buffer_cells*2 + num_center_cells

#散射区对应的格点位置
pos_center_vec   = np.arange(num_lead_cells + num_buffer_cells, num_total_cells-num_lead_cells-num_buffer_cells,1)

xn_vec   = np.arange(num_total_cells)
t_vec    = np.arange(0, 401, 50)
```

## 求解散射态（time-independant）

```python
#入射态
k      = 1
energy = 2*(1 - np.cos(k))
speed  = 2*np.sin(k)
```

### 给定输入波函数

```python
right_vec    = np.zeros(xn_vec.size, dtype=complex)
right_vec[0] = 1/np.sqrt(speed)*(1-np.exp(2j*k)) 
```

```python
diag_vec      = 2*np.ones(xn_vec.size,dtype=complex) + potential_func(xn_vec.size) - energy
offdiag_vec = - np.ones(xn_vec.size - 1)
matrix_eq   = scipy.sparse.diags([diag_vec, offdiag_vec, offdiag_vec], [0, 1, -1])
matrix_eq   = sp.csr_matrix(matrix_eq)

#边界条件
#left
matrix_eq[0,0]   = 1
matrix_eq[0,1]   = -np.exp(1j*k)
#right
matrix_eq[-1,-1] = 1
matrix_eq[-1,-2] = -np.exp(1j*k)
```

```python
psi_st_vec   = spsolve(matrix_eq, right_vec)
```

```python
plt.plot(psi_st_vec[pos_center_vec].real)
```

## 二、构建哈密顿量（包括虚势，边界无限高势磊）


### 2.1 虚势

```python
sigma_degree   = 6
sigma_strength = 33.176446627131625
sigma_vec      = -1j*(sigma_degree + 1) * sigma_strength * (np.arange(num_lead_cells)/num_lead_cells)**sigma_degree / num_lead_cells
```

```python
# Hamiltonian matrix.
diag_vec    = 2 * np.ones(num_buffer_cells*2 + num_center_cells)
diag_vec    = np.concatenate( ((sigma_vec+2*np.ones(sigma_vec.size))[::-1], diag_vec) )
diag_vec    = np.concatenate( (diag_vec, sigma_vec+2*np.ones(sigma_vec.size)) )

offdiag_vec = - np.ones(len(xn_vec) - 1)
H0          = scipy.sparse.diags([diag_vec, offdiag_vec, offdiag_vec], [0, 1, -1])
H0          = sp.csr_matrix(H0)

plt.plot(diag_vec.real)
plt.plot(diag_vec.imag)
```

### 2.2 加入电压

```python
# add the voltage pulse
def gaussianVoltage_phase_func(time):
    return 1.57 * (1 + erf((time - 100) / 20))

def hamitonian_addVoltage_func(time):
    phase        = gaussianVoltage_phase_func(time)
    #电压相位对应的hopping位置
    center_site = pos_center_vec[0]
    lead_site   = center_site - 1
    H                         = H0.copy()
    H[lead_site, center_site] = H[lead_site, center_site] * np.exp(-1j*phase)
    H[center_site, lead_site] = H[center_site, lead_site] * np.exp( 1j*phase)
    return H
```

## 三、求解所有时间点的波函数

```python
def dfdt_func(t, psi_bar_vec):
    H = hamitonian_addVoltage_func(t)
    return  ( (H-scipy.sparse.diags(energy*np.ones(num_total_cells)))@psi_bar_vec + (H - H0) @psi_st_vec)*-1j
```

```python
# %% Solve differential equation
sol = solve_ivp(dfdt_func, [t_vec[0], t_vec[-1]], np.zeros(num_total_cells,dtype=complex), t_eval=t_vec, rtol=1e-4,atol=1e-7) 
```

## 四、结果绘图

### 1. 查看中心区的波函数

```python
%matplotlib inline
# Loop over the timesteps and plot the result.
for n_t in range(t_vec.size):
    time        = t_vec[n_t]
    psi_bar_vec = sol.y[:,n_t]
    psi_vec     = (psi_bar_vec + psi_st_vec) * np.exp(-1j*energy*(time-0))
    density_vec = np.real(psi_vec * psi_vec.conjugate())
    # Prefactor and shift for representation purpose.
    plt.plot(xn_vec[pos_center_vec], 180 * density_vec[pos_center_vec] + time)
```

### 2. 查看全部区域波函数

```python
# Loop over the timesteps and plot the result.
for n_t in range(t_vec.size):
    time        = t_vec[n_t]
    psi_bar_vec = sol.y[:,n_t]
    psi_vec     = (psi_bar_vec + psi_st_vec) * np.exp(-1j*energy*(time-0))
    density_vec = np.real(psi_vec * psi_vec.conjugate())
    # Prefactor and shift for representation purpose.
    plt.plot(xn_vec, 180 * density_vec + time)
```
